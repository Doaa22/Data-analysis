#### Droping

```python
data[['gender','Partner','Churn']]
```

```python
data.drop("customerID",axis=1)  #,inplace=True)
```

```python
data.head()
```

```python
data.drop(1)
```

```python
data['gender'].replace("Male",0,inplace=True)
data['gender'].replace("Female",1,inplace=True)
data.head()
```

```python
data['gender'].value_counts() 
```

```python
def clean_id(text):
    return text.split("-")[0]
     
```

```python
data.loc[:5]
```

```python
data.loc[:2,'customerID'] = data.loc[:2,'customerID'].apply(clean_id)
data.head()
```

```python
data.head()
```

```python
def offer(row):
    if row['gender']=="Male" and row['SeniorCitizen']==1:
        return "Male_senior"
    elif row['gender']=="Male" and row['SeniorCitizen']==0:
        return "Male_junior"
    elif row['gender']=="Female" and row['SeniorCitizen']==1:
        return "Female_senior"
    elif row['gender']=="Female" and row['SeniorCitizen']==0:
        return "Female_junior"
    else:
        return "Other"
```

```python
data['offer']=data.apply(offer,axis=1)
```
